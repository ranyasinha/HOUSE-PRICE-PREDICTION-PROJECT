#Generating dataset for house price prediction

```python
import pandas as pd
import numpy as np

# Reproducibility
np.random.seed(42)

# Number of rows
n = 500

# Generate features
rooms = np.random.randint(1, 6, n)
area = np.random.randint(500, 3500, n)
bathrooms = np.random.randint(1, 4, n)
age = np.random.randint(1, 30, n)
location = np.random.choice(['Mumbai', 'Bangalore', 'Delhi'], n)

# Base price by location
location_price = {
    'Mumbai': 10000,
    'Bangalore': 8000,
    'Delhi': 7000
}

# Price formula (non-linear + noise)
price = (
    area * np.vectorize(location_price.get)(location) +
    rooms * 200000 +
    bathrooms * 150000 -
    age * 50000 +
    np.random.normal(0, 500000, n)  # noise
)

# Build dataframe
df = pd.DataFrame({
    'Rooms': rooms,
    'Area': area,
    'Bathrooms': bathrooms,
    'Age': age,
    'Location': location,
    'Price': price.astype(int)
})

# Save CSV
df.to_csv("houses.csv", index=False)

print("Dataset generated successfully: houses.csv")
print(df.head())
```
